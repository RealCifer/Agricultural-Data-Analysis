```python
import pandas as pd
import matplotlib.pyplot as plt

ka_rice_df = pd.read_csv('../data/karnataka_rice_2000_2017.csv')

ka_rice_df.head()

```

```python
plt.figure(figsize=(10,5))
plt.plot(ka_rice_df['Year'], ka_rice_df['Yield_kg_per_ha'], marker='o')
plt.xlabel('Year')
plt.ylabel('Yield (kg per ha)')
plt.title('Rice Yield Trend in Karnataka (20002017)')
plt.grid(True)
plt.show()
```
